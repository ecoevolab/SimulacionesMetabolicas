#!/bin/bash
#SBATCH --job-name=carveme
#SBATCH --output=%x.log
#SBATCH --error=%x.error
#SBATCH --time=240:00:00
#SBATCH --cpus-per-task=8
#SBATCH --mem=8G

# Info del script
date
echo "===== Beginning CarveMe pipeline ====="

# Activar ambiente conda
eval "$(conda shell.bash hook)"
module load anaconda3/2025.06
source activate carveme-1.6.6

# Directorios
INPUT_DIR="/mnt/data/sur/users/mmontante/data/rizo_data/raw_data"
OUTPUT_DIR="/mnt/data/sur/users/mmontante/data/rizo_data/carveme_prokka"

mkdir -p "$OUTPUT_DIR"

# Procesar archivos .faa (proteínas)
echo "Reconstruyendo modelos"
for file in "$INPUT_DIR"/*.faa; do
    
    base=$(basename "$file" .faa)
    echo "Modelo metabólico de: $base"
    
    carve "$file" --gapfill LB \
        --output "$OUTPUT_DIR/${base}.xml" \
        || echo "Error procesando $base"
done

echo "Reconstruyendo modelo metabolico"
for file in "$INPUT_DIR"/*.fna; do
    
    base=$(basename "$file" .fna)
    echo "Modelo metabólico para: $base"
    
    carve --dna "$file" --gapfill LB \
        --output "$OUTPUT_DIR/${base}.xml" \
        || echo "Error procesando $base"
done

echo "===== Pipeline completado ====="
date
